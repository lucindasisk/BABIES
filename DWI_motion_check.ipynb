```python
import pandas as pd
import os
from nipype.interfaces.fsl import MotionOutliers
import glob
```

```python
sixmonth = '/Users/myelin/Desktop/BABIES_motionprac'
newborn = 'Path to Newborn Data'
```

```python
sublist = []
ids = []
def get_subs(timepoint):
    subpaths = glob.glob(timepoint + '/*-*-*')
    for sub in subpaths:
        sub1 = sub.replace(timepoint+'/', '')
        ids.append(sub1)
        sub2 = sub1.replace('-C-T1', '').replace('-L-T2', '').replace('-L-T3', '')
        sublist.append(sub2)
    return print('All subs appended extracted for timepoint')
get_subs(sixmonth)
```

```python
def get_mot(sub):
    mot = MotionOutliers(in_file = timepoint +'/'+ sub + '/dwi/DTI_unwarped.nii.gz',
                        metric = metricv,
                        threshold = 0.25,
                        out_metric_values = timepoint + '/'+ sub + '/{}_motion_summary_{}.txt'.format(sub, metricv),
                        out_metric_plot = timepoint + '/'+sub +'/{}_motion_plot_{}.png'.format(sub, metricv),
                        no_motion_correction = False)
    return mot.run()
```

```python
metricv = 'dvars'
timepoint = sixmonth
for sub in ids:
    get_mot(sub)
```

```python
for sub in ['124-C-T1']:
    fd = pd.read_csv(timepoint + '/'+ sub + '/{}_motion_summary_fdrms.txt'.format(sub, metricv), header = None)
    dvars = pd.read_csv(timepoint + '/'+ sub + '/{}_motion_summary_dvars.txt'.format(sub, metricv), header = None)

dvar_math = fd - dvars
dvar_math
```
