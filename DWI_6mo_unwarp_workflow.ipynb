```python
import pandas as pd
from nibabel import load
from numpy import shape
import os
import glob
from shutil import copyfile
import nipype.pipeline.engine as pe
from nipype.interfaces import fsl
```

```python
#Set filepaths and variables
fp = '/Volumes/iang/active/BABIES/BABIES_Crossectional/BABIES_Crossectional-T1'
home = '/Volumes/iang/active/BABIES/BABIES_diffusion/subjsDir/'
destfp = '/Volumes/iang/active/BABIES/BABIES_diffusion/subjsDir/6-month'
params = home + 'acq_params.txt'
index = home + 'index.txt'
index104 = home + 'index104.txt'
index105 = home + 'index105.txt'
index1 = pd.read_csv(home + 'index104.txt', sep = " ", header = None)
workflow_dir = '/Volumes/iang/active/BABIES/BABIES_diffusion/workflows'
print(len(index1))

#Extract subject list and create new list of subjects to be processed
sub = pd.Series(glob.glob(str(fp) + '/*C-T1')).sort_values().str.lstrip('/Volumes/iang/active/BABIES/BABIES_Crossectional/BABIES_Crossectional-T')
subs = sub.str[2:12]
ready_to_run = ['100-C-T1', '102-C-T1']
```

```python
#Run through subject list and determine
for sub in subs:
    if os.path.exists(fp + '/' + sub + '/anat/dwi/recon/DTI_pe0_ms103.nii.gz'):
        print('Data found for {}'.format(sub))
        if os.path.exists(destfp + '/' + sub + '/DTI_unwarped_eddy.nii.gz'):
                print('Unwarping and eddy correction already completed for {}'.format(sub))
        else:
            print('Checking directories for {}'.format(sub))
            if os.path.exists(destfp + '/' +sub):
                print('Subject folder created for {}'.format(sub))
                if os.path.exists(destfp + '/' +sub+'/raw'):
                    print('Raw folder created for {}'.format(sub))
                    if os.path.exists(destfp + '/' +sub+'/raw/DTI_pe0_ms103.nii.gz'):
                        print('pe0 copied over')
                        if os.path.exists(destfp + '/' +sub+'/raw/DTI_pe1.nii.gz'):
                            print('pe1 also copied over')
                            ready_to_run.append(sub)
                        else:
                            copyfile(fp + '/' + sub + '/anat/dwi/recon/DTI_pe0_ms103.nii.gz', destfp + '/' + sub + '/raw/DTI_pe0_ms103.nii.gz')
                        copyfile(fp + '/' + sub + '/anat/dwi/recon/DTI_pe0_ms103.bval', 
                                 destfp + '/' + sub + '/raw/DTI_pe0_ms103.bval')
                        copyfile(fp + '/' + sub + '/anat/dwi/recon/DTI_pe0_ms103.bvec', 
                                 destfp + '/' + sub + '/raw/DTI_pe0_ms103.bvec')
                        copyfile(fp + '/' + sub + '/anat/dwi/recon/DTI_pe1.bval', 
                                 destfp + '/' + sub + '/raw/DTI_pe1.bval')
                        copyfile(fp + '/' + sub + '/anat/dwi/recon/DTI_pe1.bvec', 
                                 destfp + '/' + sub + '/raw/DTI_pe1.bvec')
                    else:
                        copyfile(fp + '/' + sub + '/anat/dwi/recon/DTI_pe0_ms103.nii.gz', destfp + '/' + sub + '/raw/DTI_pe0_ms103.nii.gz')
                        copyfile(fp + '/' + sub + '/anat/dwi/recon/DTI_pe0_ms103.bval', 
                                 destfp + '/' + sub + '/raw/DTI_pe0_ms103.bval')
                        copyfile(fp + '/' + sub + '/anat/dwi/recon/DTI_pe0_ms103.bvec', 
                                 destfp + '/' + sub + '/raw/DTI_pe0_ms103.bvec')
                        copyfile(fp + '/' + sub + '/anat/dwi/recon/DTI_pe1.nii.gz', destfp + '/' + sub + '/raw/DTI_pe1.nii.gz')
                        copyfile(fp + '/' + sub + '/anat/dwi/recon/DTI_pe1.bval', 
                                 destfp + '/' + sub + '/raw/DTI_pe1.bval')
                        copyfile(fp + '/' + sub + '/anat/dwi/recon/DTI_pe1.bvec', 
                                 destfp + '/' + sub + '/raw/DTI_pe1.bvec')
                else:
                    os.mkdir(destfp + '/' +sub+'/raw')
                    copyfile(fp + '/' + sub + '/anat/dwi/recon/DTI_pe0_ms103.nii.gz', destfp + '/' + sub + '/raw/DTI_pe0_ms103.nii.gz')
                    copyfile(fp + '/' + sub + '/anat/dwi/recon/DTI_pe0_ms103.bval', 
                             destfp + '/' + sub + '/raw/DTI_pe0_ms103.bval')
                    copyfile(fp + '/' + sub + '/anat/dwi/recon/DTI_pe0_ms103.bvec', 
                             destfp + '/' + sub + '/raw/DTI_pe0_ms103.bvec')
                    copyfile(fp + '/' + sub + '/anat/dwi/recon/DTI_pe1.nii.gz', destfp + '/' + sub + '/raw/DTI_pe1.nii.gz')
                    copyfile(fp + '/' + sub + '/anat/dwi/recon/DTI_pe1.bval', 
                             destfp + '/' + sub + '/raw/DTI_pe1.bval')
                    copyfile(fp + '/' + sub + '/anat/dwi/recon/DTI_pe1.bvec', 
                             destfp + '/' + sub + '/raw/DTI_pe1.bvec')
            else:
                os.mkdir(destfp + '/' +sub)
                os.mkdir(destfp + '/' +sub+'/raw')
                copyfile(fp + '/' + sub + '/anat/dwi/recon/DTI_pe0_ms103.nii.gz', destfp + '/' + sub + '/raw/DTI_pe0_ms103.nii.gz')
                copyfile(fp + '/' + sub + '/anat/dwi/recon/DTI_pe0_ms103.bval', 
                         destfp + '/' + sub + '/raw/DTI_pe0_ms103.bval')
                copyfile(fp + '/' + sub + '/anat/dwi/recon/DTI_pe0_ms103.bvec', 
                         destfp + '/' + sub + '/raw/DTI_pe0_ms103.bvec')
                copyfile(fp + '/' + sub + '/anat/dwi/recon/DTI_pe1.nii.gz', destfp + '/' + sub + '/raw/DTI_pe1.nii.gz')
                copyfile(fp + '/' + sub + '/anat/dwi/recon/DTI_pe1.bval', 
                         destfp + '/' + sub + '/raw/DTI_pe1.bval')
                copyfile(fp + '/' + sub + '/anat/dwi/recon/DTI_pe1.bvec', 
                         destfp + '/' + sub + '/raw/DTI_pe1.bvec')
    else:
        print('Original data not found for {}'.format(sub))
ready_to_run
```

```python
for sub in ready_to_run:
    dest = destfp + '/' + sub + '/'
    def check_bvecs():
        bvec_pe0 = pd.read_csv(dest + 'raw/DTI_pe0_ms103.bvec', sep = ' ', header = None)
        nifti_pe0 = load(dest + 'raw/DTI_pe0_ms103.nii.gz')
        index1 = pd.read_csv(home + 'index105.txt', sep = ' ', header = None)
        bv1 = len(bvec_pe0.columns)
        in1 = len(index1)
        ni1 = nifti_pe0.shape
        return print('SUBID {}: Number of entries in bvec: {}, length of index: {}, nifti dimensions: {}'.format(sub, bv1, in1, ni1))
        
    check_bvecs()
```

```python
for sub in ready_to_run:
    dest = destfp + '/' + sub + '/'
    pe1_nii = destfp + '/' + sub + '/raw/DTI_pe1.nii.gz'
    pe1_bvec = destfp + '/' + sub + '/raw/DTI_pe1.bvec'
    pe1_bval = destfp + '/' + sub + '/raw/DTI_pe1.bval'
    pe0_nii = destfp + '/' + sub + '/raw/DTI_pe0_ms103.nii.gz'
    pe0_bvec = destfp + '/' + sub + '/raw/DTI_pe0_ms103.bvec'
    pe0_bval = destfp + '/' + sub + '/raw/DTI_pe0_ms103.bval'
    fslroi = pe.MapNode(interface = fsl.ExtractROI(t_min = 0, t_size = 1),
                       name = 'fslroi',
                       iterfield = ["in_file"])
    fslroi.inputs.in_file = [pe1_nii, pe0_nii]
    fslroi.outputs.roi_file = [dest + 'pe1_b.nii.gz', dest + 'pe0_b.nii.gz']
    fslmerge = pe.Node(interface = fsl.Merge(dimension = 't', merged_file = dest + 'bud.nii.gz'), name = 'fslmerge')
    topup = pe.Node(interface = fsl.TOPUP(encoding_file = params, out_base = dest + 'DTI_topup'),
                    name = 'topup')
    dest = destfp + '/' + sub + '/'
    anatflow_c = pe.Workflow(name = "anatflow_c")
    anatflow_c.connect([(fslroi, fslmerge, [('roi_file', 'in_files')]),
                        (fslmerge, topup, [('merged_file', 'in_file')])
                       ])
    anatflow_c.base_dir = workflow_dir
    anatflow_c.write_graph(graph2use = 'flat')
    anatflow_c.run('MultiProc', plugin_args={'n_procs': 1})
ready_to_run   
```

```python
for sub in ready_to_run:
    dest = destfp + '/' + sub + '/'
    pe1_nii = destfp + '/' + sub + '/raw/DTI_pe1.nii.gz'
    pe1_bvec = destfp + '/' + sub + '/raw/DTI_pe1.bvec'
    pe1_bval = destfp + '/' + sub + '/raw/DTI_pe1.bval'
    pe0_nii = destfp + '/' + sub + '/raw/DTI_pe0_ms103.nii.gz'
    pe0_bvec = destfp + '/' + sub + '/raw/DTI_pe0_ms103.bvec'
    pe0_bval = destfp + '/' + sub + '/raw/DTI_pe0_ms103.bval'
    applytopup = pe.Node(interface = fsl.ApplyTOPUP(encoding_file = params,
                                                    in_files = [pe0_nii],
                                                    method = 'jac',
                                                    in_topup_fieldcoef = dest + 'DTI_topup_fieldcoef.nii.gz',
                                                    in_topup_movpar = dest + 'DTI_topup_movpar.txt',
                                                    in_index = [1],
                                                    out_corrected = dest + 'DTI_pe0_unwarped.nii.gz',
                                                    output_type = "NIFTI_GZ"),
                            name = 'applytopup')
    #applytopup.outputs.out_corrected = dest + 'DTI_pe0_corrected.nii.gz'
    bet = pe.Node(interface = fsl.BET(mask = True, out_file = dest + 'DTI_pe0_unwarped_stripped.nii.gz'),
                  name = 'bet')
    init_proc_c=pe.Workflow(name="init_proc_c")
    init_proc_c.connect([(applytopup, bet, [('out_corrected', 'in_file')])
                       ])
    init_proc_c.base_dir = workflow_dir
    init_proc_c.write_graph(graph2use = 'flat')
    init_proc_c.run('MultiProc', plugin_args={'n_procs': 1})
```
